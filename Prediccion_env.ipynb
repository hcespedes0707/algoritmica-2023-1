```python
df.Attrition_Flag =df.Attrition_Flag.map({'Existing Customer':0,'Attrited Customer':1})
```

```python
df.Gender =df.Gender.map({'M':0,'F':1})
```

```python
df.Education_Level =df.Education_Level.map({'High School':0, 'Graduate':1, 'Uneducated':2, 'College':3,
       'Post-Graduate':4, 'Doctorate':5})
```

```python
df.Marital_Status =df.Marital_Status.map({'Married':0, 'Single':1, 'Divorced':2})
```

```python
df.Income_Category =df.Income_Category.map({'$60K - $80K':0, 'Less than $40K':1, '$80K - $120K':2, '$40K - $60K':3,
       '$120K +':4})
```

```python
df.Card_Category.unique()
```

```python
df.Card_Category =df.Card_Category.map({'Blue':0, 'Gold':1, 'Silver':2,'Platinum':3})
```

```python
data_x= df.drop(['CLIENTNUM','Attrition_Flag','Customer_Age','Gender','Dependent_count','Education_Level','Marital_Status','Income_Category','Months_on_book','Total_Relationship_Count','Months_Inactive_12_mon','Contacts_Count_12_mon','Total_Revolving_Bal','Avg_Open_To_Buy','Total_Trans_Amt','Total_Trans_Ct','Total_Ct_Chng_Q4_Q1','Total_Amt_Chng_Q4_Q1','Avg_Utilization_Ratio','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'],axis=1)
data_y = df.Avg_Open_To_Buy
```

```python
train_x,test_x,train_y,test_y = train_test_split(data_x,data_y,test_size=0.1)
```

```python
LinearRegression=linear_model.LinearRegression()
```

```python
LinearRegression.fit(train_x,train_y)
```

```python
y_predicted = LinearRegression.predict(test_x)
```

```python
print(f"La ecuacion de la recta es y ={round(LinearRegression.coef_[0],2)}x + {round(LinearRegression.intercept_,2)}")
```

```python
print(f"El score en prediccion es:{LinearRegression.score(test_x, test_y)}")
```

```python
print(f'Score: {round(LinearRegression.score(test_x, test_y), 4) * 100}%')
```

```python
Limite_Credito = 1500
card_categoria = 2 

predicted = LinearRegression.predict([[Limite_Credito, card_categoria]])
labels = {0:'Blue', 1: 'Gold',2: 'Silver',3: 'Platinum'}
```

```python
print(f'El cliente con un limite de credito {Limite_Credito}, que es considerado categoria {labels[card_categoria]}  aproximadamente ha realiado  al banco de  {round(predicted[0], 2)} Bolivianos')
```
